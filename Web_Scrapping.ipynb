```python
import csv
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options

options=Options()
options.headless=True
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

def scrap_reviews_from_page(url):
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")

    restaurant_name = soup.find('h1', class_='xbVaJ4QYPgE-').get_text(strip=True)

    reviews = []
    review_container = soup.find_all('div', class_='MpiILQAMSSg-')
    for review in review_container:
        review_text = review.find('div', class_='_6rFG6U7PA6M-').get_text(strip=True)
        review_rating = review.find('li', class_='-k5xpTfSXac-').get_text(strip=True)

        reviews.append({
            'restaurant_name': restaurant_name,
            'review_text': review_text,
            'review_rating': review_rating
        })
    
    return reviews

def scrap_all_reviews(url, max_page=161):
    all_reviews = []
    
    for page_num in range(1, max_page + 1):  # Loop through pages 1 to 161
        current_url = f"{url}&page={page_num}"
        print(f"Scraping page {page_num}: {current_url}")
        
        reviews = scrap_reviews_from_page(current_url)
        
        all_reviews.extend(reviews)
        
        time.sleep(2)  # Wait for 2 seconds between requests

    return all_reviews


def save_reviews_to_csv(reviews, filename='reviews.csv'):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer=csv.writer(file)

        writer.writerow(['restaurant_name', 'review_text', 'review_rating'])

        for review in reviews:
            writer.writerow([review['restaurant_name'], review['review_text'], review['review_rating']])
    
    print(f"Reviews saved to {filename}")


# URL for the first page
url = 'https://www.opentable.com/gary-danko?originId=3d44b960-7286-4e34-9b1a-2dd20584062d&corrid=3d44b960-7286-4e34-9b1a-2dd20584062d&avt=eyJ2IjoyLCJtIjoxLCJwIjowLCJzIjowLCJuIjowfQ'

all_reviews = scrap_all_reviews(url)
print(f"\nTotal reviews scraped: {len(all_reviews)}")

save_reviews_to_csv(all_reviews)

driver.quit()
```
